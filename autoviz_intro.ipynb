```python
from autoviz import AutoViz_Class
import pandas as pd
import numpy as np
```

```python
AV = AutoViz_Class()
%matplotlib inline
filename = "data/Housing.csv"
```

```python
df = AV.AutoViz(filename)
```

```python
### FixDQ() Function
```

```python
from autoviz import FixDQ
fixdq = FixDQ()
```

```python
fixdq.detect_duplicates(df)
```

```python
fixdq.fit_transform(df)
```

```python
### Using a Target Variable and More Arguments
```

```python
target_variable = "price"

dft = AV.AutoViz(
    filename,
    sep=",",
    depVar=target_variable,
    dfte=None,
    header=0, #default
    verbose=1, #default
    lowess=True,
    chart_format="svg", #default
    max_rows_analyzed=150000, #default
    max_cols_analyzed=30, #default
    save_plot_dir=None,
)
```
